```python
import pandas as pd
import numpy as np
```

# Cubesat morning 

```python
cube = pd.read_csv('March18_Morning_data.csv')
weather = pd.read_csv('weather_data_March18.csv')
sm_weather = pd.read_csv('ambient-weather-stone.csv')
```

```python
cube
```

```python
cube['timestamp'] = pd.to_datetime(cube['timestamp'])
cube.sort_values(by='timestamp', ascending=True, inplace=True)
cube['timestamp'] = cube['timestamp'].dt.floor('Min') 
cube = cube.resample('1Min', on='timestamp').mean()
cube.dropna(inplace=True)
cube.head()
```

```python
weather.reset_index(inplace=True)
weather['SimpleDate'] = pd.to_datetime(weather['SimpleDate'])
weather.sort_values(by='SimpleDate', ascending=True, inplace=True)
weather['SimpleDate'] = weather['SimpleDate'].dt.floor('Min')
weather.rename(columns={'SimpleDate': 'timestamp'}, inplace=True)
weather = weather.resample('1Min', on='timestamp').mean()
```

```python
for column in weather.columns:
    cube[column] = weather[column]
cube.head()
```

```python
sm_weather.reset_index(inplace=True)
sm_weather['SimpleDate'] = pd.to_datetime(sm_weather['SimpleDate'])
sm_weather.sort_values(by='SimpleDate', ascending=True, inplace=True)
sm_weather['SimpleDate'] = sm_weather['SimpleDate'].dt.floor('Min')
sm_weather.rename(columns={'SimpleDate': 'timestamp'}, inplace=True)
sm_weather = sm_weather.resample('1Min', on='timestamp').mean()
sm_weather.head()
```

```python
for column in sm_weather.columns:
    cube[column] = sm_weather[column]
cube.head()
```

```python
#cube.interpolate(method='spline', inplace=True, order=3, limit_direction='both')
cube.interpolate(method='linear', inplace=True, limit_direction='both')
cube.head()
```

```python
cube.dropna(inplace=True)
cube.reset_index(inplace=True)
```

```python
cube.to_csv('cube_run_morning_combined.csv', index=False)
```
