```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#import the countries data from the csv file with an index column of countries
df_morn = pd.read_csv("cube_run_morning_combined.csv", index_col="timestamp")
df_morn.info()

#create columns for counts
counts_1_2 = df_morn['Chan_0_1']
counts_1_3 = df_morn['Chan_0_2']
counts_2_3 = df_morn['Chan_1_2']
```

```python
df_morn.dropna(inplace=True)
```

```python
df_morn.info()
```

```python
df_morn['average'] = df_morn[['Chan_0_1', 'Chan_0_2', 'Chan_1_2']].mean(axis=1)
df_morn.head()
```

```python
#plot the figure with multiple axes
fig, ax1 = plt.subplots(figsize=(20,10))
#plot average counts on the first axis
ax1.plot(df_morn.index, df_morn['average'], color='blue')
ax1.set_xlabel('Time')
#ax1.set_ylabel('Average Counts', color='blue')
#plot humidity on the second axis
ax1.plot(df_morn.index, df_morn['Humidity'], color='red')
#ax2.set_ylabel('Humidity', color='red')
#plot temperature on the third axis
ax1.plot(df_morn.index, df_morn['Temperature'], color='green')
#ax3.set_ylabel('Temperature', color='green')
#plot pressure on the fourth axis
ax1.plot(df_morn.index, df_morn['Relative Pressure'], color='orange')
ax1.plot(df_morn.index, df_morn['Humidity_SM'], color='purple')
ax1.plot(df_morn.index, df_morn['Temperature_SM'], color='pink')
ax1.plot(df_morn.index, df_morn['RelativePressure_SM'], color='black')
#ax4.set_ylabel('Pressure', color='orange')
#set the title
ax1.set_title('Morning Counts and Weather Data')
#set the legend
ax1.legend(['Average Counts', 'Atlanta Humidity', 'Atlanta Temperature', 'Atlanta Pressure', 'Stone Mountain Humidity', 'Stone Mountain Temperature', 'Stone Mountain Pressure'], loc='upper left')
```

```python
df_atlanta = pd.read_csv('muon001_hourly_alldata.csv',sep='\s+')
#combine index with timestamp column to get a datetime index
df_atlanta.index = df_atlanta.index + ' ' + df_atlanta['timestamp']
#drop the timestamp column
df_atlanta.drop(columns=['timestamp'], inplace=True)
df_atlanta.index = pd.to_datetime(df_atlanta.index)
#only keep datetimes after 2023-01-01
df_atlanta = df_atlanta[df_atlanta.index > '2023-01-01']
df_atlanta.head()
```

```python
df_atlanta['average_m_avg'] = df_atlanta[['Ch12_m_avg', 'Ch13_m_avg', 'Ch23_m_avg']].mean(axis=1)
df_atlanta.head()
```

```python
#plot the figure with multiple axes
fig, ax1 = plt.subplots()
#plot average counts on the first axis
ax1.plot(df_atlanta.index, df_atlanta['average_m_avg'], color='blue')
ax1.plot(df_atlanta.index, df_atlanta['temp_scaled'], color='orange')
ax1.plot(df_atlanta.index, df_atlanta['press_pct'], color='green')
```
