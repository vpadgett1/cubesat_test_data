```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# Load the data into a Pandas DataFrame
data = pd.read_csv('March18_Morning_data.csv', parse_dates=['timestamp'])
lat_min = data['Latitude'].min()
lat_max = data['Latitude'].max()
lon_min = data['Longitude'].min()
lon_max = data['Longitude'].max()
# Create a Basemap object for the region of interest
m = Basemap(llcrnrlon=lon_min-0.05, llcrnrlat=lat_min-0.05, urcrnrlon=lon_max+0.05, urcrnrlat=lat_max+0.05, resolution='h')

# Plot the map
m.drawcoastlines()
m.drawcountries()
m.drawstates()  

# Convert latitude and longitude to map coordinates
x, y = m(data['Longitude'].values, data['Latitude'].values)

df_aveg_counts = data[['Chan_0_1', 'Chan_0_2', 'Chan_1_2']].mean(axis=1)
df_aveg_counts.head()

figure, axes = plt.subplots(2, 2, figsize=(10, 10))
axes[0, 0].scatter(x, y, s=data['Chan_0_1'], c='red', alpha=0.5)
axes[0,0].set_title('Morning Channels 0 and 1 Data')
axes[0,0].set_xlabel('Longitude')
axes[0,0].set_ylabel('Latitude')
axes[0, 1].scatter(x, y, s=data['Chan_0_2'], c='green', alpha=0.5)
axes[0,1].set_title('Morning Channels 0 and 2 Data')
axes[0,1].set_xlabel('Longitude')
axes[0,1].set_ylabel('Latitude')
axes[1, 0].scatter(x, y, s=data['Chan_1_2'], c='blue', alpha=0.5)
axes[1,0].set_title('Morning Channels 1 and 2 Data')
axes[1,0].set_xlabel('Longitude')
axes[1,0].set_ylabel('Latitude')
axes[1, 1].scatter(x, y, s=df_aveg_counts, c='purple', alpha=0.5)
axes[1,1].set_title('Morning Average Counts Data')
axes[1,1].set_xlabel('Longitude')
axes[1,1].set_ylabel('Latitude')


# # Plot the data on the map
# plt.scatter(x, y, s=data['Chan_0_1'], c='red', alpha=0.5)
# # plt.scatter(x, y, s=data['Chan_0_2'], c='green', alpha=0.5)
# #plt.scatter(x, y, s=data['Chan_1_2'], c='blue', alpha=0.5)

# # Adjust plot parameters
# plt.title('Channel Data')
# plt.xlabel('Longitude')
# plt.ylabel('Latitude')

# # Show the plot
# plt.show()
```
