```python
receiving.sort_values(by='timestamp', ascending=True, inplace=True)
receiving.head()
```

```python
receiving = receiving[['timestamp', 'counts_1_2', 'counts_1_3', 'counts_2_3']]
```

```python
receiving = receiving.resample('H', on = 'timestamp', axis=0).mean()
receiving.head()
```

```python
receiving= receiving.interpolate(method='linear', limit_direction='both')  
receiving.drop_duplicates(inplace=True)

receiving.head(20)
```
