```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#import the countries data from the csv file with an index column of countries
df_ames = pd.read_csv("trip2NASA_Ames_Jun26to28_2023.csv")
#timestamp column to datetime 
df_ames['timestamp'] = pd.to_datetime(df_ames['timestamp'])
df_ames = df_ames.resample('H', on ='timestamp', axis=0).mean()
df_ames.head()
```

```python
#plot the figure with multiple axes
fig, ax1 = plt.subplots(figsize=(20,10))
#plot average counts on the first axis
#ax1.plot(df_combined.index, df_combined['average'], color='blue')
ax1.set_xlabel('Time')

ax1.plot(df_ames.index, df_ames['Chan_0_1'], color='blue', linestyle='dotted')
ax1.plot(df_ames.index, df_ames['Chan_0_2'], color='green', linestyle="dashed")
ax1.plot(df_ames.index, df_ames['Chan_1_2'], color='purple')
#set the title
ax1.set_title('June 26th to June 28th 2023 Flight to Mountain View, CA')
#set the legend
ax1.legend(['Channel_0_1', 'Channel_0_2', 'Channel_1_2'], loc='upper left')
```

```python
#make zero mean and unit variance
df_ames_zmuv = df_ames.copy()
df_ames_zmuv['Chan_0_1'] = (df_ames_zmuv['Chan_0_1'] - df_ames_zmuv['Chan_0_1'].mean()) / df_ames_zmuv['Chan_0_1'].std()
df_ames_zmuv['Chan_0_2'] = (df_ames_zmuv['Chan_0_2'] - df_ames_zmuv['Chan_0_2'].mean()) / df_ames_zmuv['Chan_0_2'].std()
df_ames_zmuv['Chan_1_2'] = (df_ames_zmuv['Chan_1_2'] - df_ames_zmuv['Chan_1_2'].mean()) / df_ames_zmuv['Chan_1_2'].std()
df_ames_zmuv.head()
```

```python
#plot the figure with multiple axes
fig, ax1 = plt.subplots(figsize=(20,10))
#plot average counts on the first axis
#ax1.plot(df_combined.index, df_combined['average'], color='blue')
ax1.set_xlabel('Time')

ax1.plot(df_ames_zmuv.index, df_ames_zmuv['Chan_0_1'], color='blue', linestyle='dotted')
ax1.plot(df_ames_zmuv.index, df_ames_zmuv['Chan_0_2'], color='green', linestyle="dashed")
ax1.plot(df_ames_zmuv.index, df_ames_zmuv['Chan_1_2'], color='purple')
#set the title
ax1.set_title('June 26th to June 28th 2023 Flight to Mountain View, CA')
#set the legend
ax1.legend(['Channel_0_1', 'Channel_0_2', 'Channel_1_2'], loc='upper left')
```
