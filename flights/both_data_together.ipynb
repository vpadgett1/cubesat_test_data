```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#import the countries data from the csv file with an index column of countries
df_india_arrival = pd.read_csv("india_flight_there_data.csv")
#timestamp column to datetime 
df_india_arrival['timestamp'] = pd.to_datetime(df_india_arrival['timestamp'])
df_india_arrival = df_india_arrival.resample('H', on ='timestamp', axis=0).mean()
df_india_arrival.head()
```

```python
#import the countries data from the csv file with an index column of countries
df_ames_arrival = pd.read_csv("trip2NASA_Ames_flight_there_data.csv")
#timestamp column to datetime 
df_ames_arrival['timestamp'] = pd.to_datetime(df_ames_arrival['timestamp'])
df_ames_arrival = df_ames_arrival.resample('H', on ='timestamp', axis=0).mean()
df_ames_arrival.head()
```

```python
df_india_arrival.reset_index(inplace=True)
df_india_arrival.head()
```

```python
#subtract day from timestamp india data
df_india_arrival['timestamp'] = df_india_arrival['timestamp'] - min(df_india_arrival['timestamp'])
df_india_arrival['timestamp'] = df_india_arrival['timestamp'].dt.total_seconds()
df_india_arrival['timestamp'] = df_india_arrival['timestamp'] / 3600
df_india_arrival.head()
```

```python
df_ames_arrival.reset_index(inplace=True)
df_ames_arrival.head()
```

```python
df_ames_arrival['timestamp'] = df_ames_arrival['timestamp'] - min(df_ames_arrival['timestamp'])
df_ames_arrival['timestamp'] = df_ames_arrival['timestamp'].dt.total_seconds()
df_ames_arrival['timestamp'] = df_ames_arrival['timestamp'] / 3600
df_ames_arrival.head()
```

```python
#plot the figure with multiple axes
fig, ax1 = plt.subplots(figsize=(20,10))
#plot average counts on the first axis
#ax1.plot(df_combined.index, df_combined['average'], color='blue')
ax1.set_xlabel('Time')

ax1.plot(df_india_arrival.index, df_india_arrival['Chan_0_1'], color='blue', linestyle='dotted', label='India Channel_0_1')
ax1.plot(df_india_arrival.index, df_india_arrival['Chan_0_2'], color='green', linestyle="dashed", label='India Channel_0_2')
ax1.plot(df_india_arrival.index, df_india_arrival['Chan_1_2'], color='purple', label='India Channel_1_2')

ax1.plot(df_ames_arrival.index, df_ames_arrival['Chan_0_1'], color='red', linestyle='dotted', label='Ames Channel_0_1')
ax1.plot(df_ames_arrival.index, df_ames_arrival['Chan_0_2'], color='orange', linestyle="dashed", label='Ames Channel_0_2')
ax1.plot(df_ames_arrival.index, df_ames_arrival['Chan_1_2'], color='gold', label='Ames Channel_1_2')
#set the title
#set the title
ax1.set_title('Comparison of the Departing Flights from Atlanta to India and and NASA Ames')
#set the legend
ax1.legend(loc='upper left')
```
