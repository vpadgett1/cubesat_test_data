```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#import the countries data from the csv file with an index column of countries
df_india = pd.read_csv("india_trip_data.csv")
#timestamp column to datetime 
df_india['timestamp'] = pd.to_datetime(df_india['timestamp'])
df_india = df_india.resample('H', on ='timestamp', axis=0).mean()
df_india.head()
```

```python
#plot the figure with multiple axes
fig, ax1 = plt.subplots(figsize=(20,10))
#plot average counts on the first axis
#ax1.plot(df_combined.index, df_combined['average'], color='blue')
ax1.set_xlabel('Time')

ax1.plot(df_india.index, df_india['Chan_0_1'], color='blue', linestyle='dotted')
ax1.plot(df_india.index, df_india['Chan_0_2'], color='green', linestyle="dashed")
ax1.plot(df_india.index, df_india['Chan_1_2'], color='purple')
#set the title
ax1.set_title('November 26th to November 28th and December 21st Flights To and From Atlanta to India')
#set the legend
ax1.legend(['Channel_0_1', 'Channel_0_2', 'Channel_1_2'], loc='upper left')
```

take the ratio between the ratio of difference between channels and plot one line for each channel 

```python
#make zero mean and unit variance
df_india_zmuv = df_india.copy()
df_india_zmuv['Chan_0_1'] = (df_india_zmuv['Chan_0_1'] - df_india_zmuv['Chan_0_1'].mean()) / df_india_zmuv['Chan_0_1'].std()
df_india_zmuv['Chan_0_2'] = (df_india_zmuv['Chan_0_2'] - df_india_zmuv['Chan_0_2'].mean()) / df_india_zmuv['Chan_0_2'].std()
df_india_zmuv['Chan_1_2'] = (df_india_zmuv['Chan_1_2'] - df_india_zmuv['Chan_1_2'].mean()) / df_india_zmuv['Chan_1_2'].std()
df_india_zmuv.head()
```

```python
#plot the figure with multiple axes
fig, ax1 = plt.subplots(figsize=(20,10))
#plot average counts on the first axis
#ax1.plot(df_combined.index, df_combined['average'], color='blue')
ax1.set_xlabel('Time')

ax1.plot(df_india_zmuv.index, df_india_zmuv['Chan_0_1'], color='blue', linestyle='dotted')
ax1.plot(df_india_zmuv.index, df_india_zmuv['Chan_0_2'], color='green', linestyle="dashed")
ax1.plot(df_india_zmuv.index, df_india_zmuv['Chan_1_2'], color='purple')
#set the title
ax1.set_title('June 26th to June 28th 2023 Flight to Mountain View, CA')
#set the legend
ax1.legend(['Channel_0_1', 'Channel_0_2', 'Channel_1_2'], loc='upper left')
```

```python
df_india_counts = pd.read_csv("updated_timestamp_data.csv")
df_india_counts['timestamp'] = pd.to_datetime(df_india_counts['timestamp'])
df_india_counts = df_india_counts.set_index('timestamp')
#df_india_counts = df_india_counts.resample('H', on ='timestamp', axis=0).sum()
df_india_counts.head()
```

```python
df_india_counts_copy = df_india_counts.copy()
df_india_counts_copy.reset_index(inplace=True)
df_india_counts_copy["changed_times"] = df_india_counts_copy['timestamp']
# df_india_counts_copy[df_india_counts_copy['timestamp'] > pd.to_datetime("2023-11-27 18:06:07")]["changed_times"] = df_india_counts_copy['timestamp'] - pd.Timedelta(hours=9, minutes=0, seconds=0)
# #df_india_counts_copy.head()
# df_india_counts_copy[df_india_counts_copy['timestamp'] > pd.to_datetime("2023-11-27 10:00:07")]

# Define the timestamp after which you want to subtract 9 hours
certain_timestamp = pd.Timestamp('2023-11-27 18:06:07')

# Subtract 9 hours from timestamps after the certain_timestamp
mask = df_india_counts_copy['changed_times'] > certain_timestamp
print(mask)
df_india_counts_copy.loc[mask, 'changed_times'] -= pd.Timedelta(hours=13)

print("Original DataFrame:")
print(df_india_counts_copy)

#print the first row where the changed timestamp is not the same
print(df_india_counts_copy[df_india_counts_copy['timestamp'] != df_india_counts_copy['changed_times']].head(1))

#reset changed times as index
df_india_counts_copy.set_index('changed_times', inplace=True)
```

```python
#plot dataframe above
fig, ax1 = plt.subplots(figsize=(20,10))
ax1.plot(df_india_counts_copy.index, df_india_counts_copy['Chan_0_1'], color='red', marker='.', markersize=2, linestyle='none', linewidth=0.5)
# ax1.plot(df_india_counts.index, df_india_counts['Chan_0_2'], color='green', linestyle="dashed")
# ax1.plot(df_india_counts.index, df_india_counts['Chan_1_2'], color='purple')

#set the title
ax1.set_title('Flight from Atlanta, Georgia to Delhi, India from November 26th 2023 to November 28th 2023')
#set the legend
ax1.legend(['CubeSat Channel_0_1 per Minute Counts'], loc='upper right')

#set x axis to timestamp
ax1.set_xlabel('Date and Time', loc='right')
#set y axis to counts
ax1.set_ylabel('Counts per Minute', loc="top")
```
